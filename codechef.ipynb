```python
import pandas as pd
```

```python
dataset = pd.read_csv("data.csv", header=None)
dataset
```

```python
X = dataset.drop(columns=2)
y = dataset.drop(columns=[0, 1])
```

```python
import numpy as np
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
X = np.array(X)
y = np.array(y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
mpl.rcParams['legend.fontsize'] = 12
fig = plt.figure()
ax = fig.gca(projection ='3d').scatter(X[:, 0], X[:, 1], y, label ='y', s = 5)
# ax = fig.gca(projection ='3d').scatter(X[:, 0], X[:, 1], lr.predict(X), label ='y', s = 5, color='red')
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset[0], dataset[2], color='black')
plt.scatter([X[i][0] for i in range(len(X))], lr.predict(X), color='red')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 3)
X_poly = poly.fit_transform(X)
poly.fit(X_poly, y)

lr2 = LinearRegression()
lr2.fit(X_poly, y)

# pred = lr2.predict(poly.fit_transform([[1, 1200]]))
```

```python
mpl.rcParams['legend.fontsize'] = 12
fig = plt.figure()
ax = fig.gca(projection ='3d').scatter(X[:, 0], X[:, 1], y, label ='y', s = 5)
ax = fig.gca(projection ='3d').scatter(X[:, 0],
                                       X[:, 1], 
                                       lr2.predict(poly.fit_transform(X)), 
                                       label ='y', 
                                       s = 5, 
                                       color='red')
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset[0], dataset[2], color='black')
plt.scatter([X[i][0] for i in range(len(X))], lr2.predict(poly.fit_transform(X)), color='red')
```
